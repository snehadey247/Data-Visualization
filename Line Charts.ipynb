```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
spotify_filepath = "spotify.csv"

# Read the file into a variable spotify_data
spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)
```

```python
# Print the first 5 rows of the data
spotify_data.head()
```

```python
# Print the last five rows of the data
spotify_data.tail()
```

## Plot the data 

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of each song 
sns.lineplot(data=spotify_data['Streams'])
```

## Plot a subset of the data

```python
list(spotify_data.columns)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of 'Position'
sns.lineplot(data=spotify_data['Position'], label="Position")

# Line chart showing daily global streams of 'Streams'
sns.lineplot(data=spotify_data['Streams'], label="Streams")


# Add label for horizontal axis
plt.xlabel("Date")
```
